```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
line = 'a'
kod = []
wynik = []
with open('2022.txt') as f:
  line = f.readline()
  while line != '':
    # split line into list
    line = line.split()
    # check if line elements are numbers
    if line[0].isdigit():
      # append first element to kod list as a number
      # append second element to wynik list as a number
      kod.append(int(line[0]))
      wynik.append(int(line[1]))
    line = f.readline()
    if len(line) == 0:
      break
```

```python
# load kod and wynik into dataframe
df = pd.DataFrame({'kod': kod, 'wynik': wynik})
df
```

```python
plt.figure(figsize=(15, 10))
values, bins, bars = plt.hist(df['wynik'], bins=range(0,150,5), color='green', edgecolor='black', linewidth=1.2)
plt.bar_label(bars, fontsize=10)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Wynik')
plt.ylabel('Liczba osób')
plt.title('Histogram wyników')
```
